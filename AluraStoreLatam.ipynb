```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calculate the profit for each sale in each dataframe
tienda['Ganancia'] = tienda['Precio'] - tienda['Costo de envío']
tienda2['Ganancia'] = tienda2['Precio'] - tienda2['Costo de envío']
tienda3['Ganancia'] = tienda3['Precio'] - tienda3['Costo de envío']
tienda4['Ganancia'] = tienda4['Precio'] - tienda4['Costo de envío']

# Calculate the total profit for each dataframe
total_profit_tienda1 = tienda['Ganancia'].sum()
total_profit_tienda2 = tienda2['Ganancia'].sum()
total_profit_tienda3 = tienda3['Ganancia'].sum()
total_profit_tienda4 = tienda4['Ganancia'].sum()

# Create a dictionary with store names and their total profits
profit_data = {
    'Tienda 1': total_profit_tienda1,
    'Tienda 2': total_profit_tienda2,
    'Tienda 3': total_profit_tienda3,
    'Tienda 4': total_profit_tienda4
}

# Create a pandas Series from the dictionary
profit_series = pd.Series(profit_data)

# Create a bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=profit_series.index, y=profit_series.values, palette='viridis') # Added palette
plt.title('Ganancia Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ganancia Total (Millones)')

# Format y-axis labels to show values in millions
formatter = mticker.FuncFormatter(lambda x, _: '{:.3f}M'.format(x/1e6))
ax.yaxis.set_major_formatter(formatter)

plt.show()
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd

# Combine all dataframes into a single dataframe
all_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

# Calculate the average rating for each individual store
average_rating_tienda1 = tienda['Calificación'].mean()
average_rating_tienda2 = tienda2['Calificación'].mean()
average_rating_tienda3 = tienda3['Calificación'].mean()
average_rating_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {average_rating_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {average_rating_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {average_rating_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {average_rating_tienda4:.2f}")

# Calculate the average rating for each location in the combined dataframe
average_rating_by_location = all_tiendas.groupby('Lugar de Compra')['Calificación'].mean()

# Get the top 3 locations with the highest average rating
top_3_locations = average_rating_by_location.nlargest(3)

print("\nLas 3 ubicaciones con la mejor calificación promedio:")
display(top_3_locations)
```

# 4. Productos más y menos vendidos

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Combine all dataframes into a single dataframe
all_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

# Count the occurrences of each product
product_counts = all_tiendas['Producto'].value_counts()

# Get the most and least sold products
most_sold_products = product_counts.head()
least_sold_products = product_counts.tail()

print("Productos más vendidos:")
display(most_sold_products)

print("\nProductos menos vendidos:")
display(least_sold_products)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Add a 'Tienda' column to each dataframe
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combine all dataframes into a single dataframe
all_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

# Count the occurrences of each product
product_counts = all_tiendas['Producto'].value_counts()

# Get the most and least sold products
most_sold_products = product_counts.head()
least_sold_products = product_counts.tail()

# Combine the most and least sold product lists
top_bottom_products = pd.concat([most_sold_products, least_sold_products])

# Filter the combined dataframe to include only the top and bottom products
filtered_tiendas = all_tiendas[all_tiendas['Producto'].isin(top_bottom_products.index)]

# Count the occurrences of these products by store
product_store_counts = filtered_tiendas.groupby(['Producto', 'Tienda']).size().unstack(fill_value=0)

# Create a grouped bar chart
product_store_counts.plot(kind='bar', figsize=(15, 7))
plt.title('Distribución de Productos Más y Menos Vendidos por Tienda')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```
